# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first 10 rows
df.head(10)

```

```python
# length of data 
len(df)
```

```python

# Check for missing values and data types

df.isna().sum()
df.isna().sum().sum()
df.dtypes
```

```python
df['housing'].value_counts()
print("total no of unkown values in housing")
(df['housing'] == 'unknown').sum()

#if unknown is the missing value in housing then this formula will show the number of unkown value
```

```python


df['loan'].value_counts()
print("total no of unkown values in loan")
(df['loan'] == 'unknown').sum()


    

#if unknown is the missing value in loan then this formula will show the number of unkown value
```

```python

# Summary statistics

df.describe()
```

```python
df.columns
```

```python

import matplotlib.pyplot as plt

# Plot the distribution of the target variable

df['y'].value_counts().plot(kind='bar')
plt.title("Distribution of Target Variable")
plt.show()



df.columns
```

```python

# Correlation matrix for numerical features

numfeautre =  df.select_dtypes(include=['int64', 'float64'])
#creating correltion matrix
corr_matrix = numfeautre.corr()
#print matrix
print(corr_matrix)
```

```python
#visulize correltion matrix using heatmap
import seaborn as sns


plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# a very important graph because i noticed for high duration of time customer says yes for short durtion of time customer says no.
#for improving imbalnced dataset durtion of time should increase then customer will say yes
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='y', y='duration', data=df)
plt.title("Average Duration vs Subscription")
plt.show()
```
